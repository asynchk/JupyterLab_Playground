```python

parcel_data = pd.read_csv('./parcelLetter.csv', index_col="Parcel ID")
input_parameters = ['Origin', 'Destination', 'Weight']
output_class = ['Type']
training_data = parcel_data.loc[:20, input_parameters]
training_data_result = parcel_data.loc[:20, output_class]
testing_data = parcel_data.loc[21:, input_parameters]
```

```python

def hm_d(r1, r2):
#     print(r1)
#     print(r2);
#     print(r1.ne(r2))
    
    return r1.ne(r2).sum()/r1.size
    
    

def cal_hd_to_all(row):
    hamming_distance_row = training_data.apply(lambda d: hm_d(d, row), axis="columns")
    return hamming_distance_row
#     print(row)
#     return row.sum()


hamming_distance_data = training_data.apply(cal_hd_to_all, axis="columns")


hamming_distance_data


# training_data

# training_data
# cal_hd_to_all(training_data.iloc[2])

# training_data.iloc[0] 
# training_data.iloc[0].eq(training_data.iloc[1] )

```

```python
def get_closest_points(d_data, k=3):
    sort_data = d_data.sort_values(ascending = True);
    return sort_data.index[:3]

closest3points = hamming_distance_data.apply(get_closest_points, axis='rows')
closest3points
```
